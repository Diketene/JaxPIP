```python
import os
from cqpes.data import PIPData
from jax import numpy as jnp
from jaxpip import JaxPIPDescriptor
```

```python
SiH4_data = PIPData.from_file(
    xyz_file=os.path.join("data", "SiH4.xyz"),
    E_file=os.path.join("data", "SiH4_CCSD-T.dat"),
)

SiH4_data.set_reference_energy()

print(repr(SiH4_data))
```

```python
SiH4_xyz = jnp.asarray(SiH4_data[0].xyz)
SiH4_xyz
```

```python
ab4_pip = JaxPIPDescriptor.from_json(
    os.path.join("basis", "MOL_1_4_4.json"),
    alpha=1.0,
)
```

```python
# distance vector r
r = ab4_pip.calc_r_from_xyz(SiH4_xyz)
r
```

```python
# morse vector m
m = ab4_pip.calc_m_from_r(r)
m
```
