```python
import os
import dpdata

import jax

# jax.config.update("jax_enable_x64", True)

from jax import numpy as jnp
from jaxpip.descriptor import PIPDescriptor
```

```python
# load CH4.xyz
s = dpdata.System("CH4.xyz", fmt="xyz")
s
```

```python
# show xyz
xyz = jnp.array(s.data["coords"][0])
xyz
```

```python
a4b_pip = PIPDescriptor.from_json(
    os.path.join("basis", "A4B_4.json"),
    alpha=1.0,
    with_grad=False,
    dtype=jnp.float32,
)

a4b_pip
```

```python
a4b_pip_with_grad = PIPDescriptor.from_json(
    os.path.join("basis", "A4B_4.json"),
    alpha=1.0,
    with_grad=True,
    dtype=jnp.float32,
)

a4b_pip_with_grad
```

```python
p = a4b_pip(xyz)

p
```

```python
%%timeit

p = a4b_pip(xyz)
```

```python
p, J_p_xyz = a4b_pip_with_grad(xyz)
```
