```python
import os
from cqpes.data import PIPData
from jax import numpy as jnp
from jaxpip.descriptor import PIPDescriptor
```

```python
# load CH4.xyz
data = PIPData.from_file("CH4.xyz")
```

```python
# show xyz
xyz = jnp.array(data[0].xyz)
xyz
```

```python
# initialize PIP descriptor

a4b_pip = PIPDescriptor.from_json(
    os.path.join("basis", "A4B_4.json")
)

a4b_pip
```

```python
# calculate PIP and its Jacobian matrix with respect to xyz
p, J_p_xyz = a4b_pip(xyz, with_grad=True)
```
