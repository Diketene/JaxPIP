```python
import gzip
import json
import os
import time

import jax
from jax import numpy as jnp

from jaxpip.descriptor import PIPDescriptor
```

```python
_DTYPE = jnp.float32
```

```python
key = jax.random.PRNGKey(114514)
```

```python
basis_file = os.path.join("basis", "MOL_5_2_1_7.json.gz")

with gzip.open(basis_file, mode="rt") as f:
    basis_set = json.load(f)

len(basis_set)
```

```python
pip = PIPDescriptor(
    basis_set=basis_set,
    alpha=1.0,
    with_grad=False,
    dtype=_DTYPE,
)
```

```python
pip_with_grad = PIPDescriptor(
    basis_set=basis_set,
    alpha=1.0,
    with_grad=True,
    dtype=_DTYPE,
)
```

```python
xyz = jax.random.normal(key=key, shape=(pip.num_atoms, 3))

xyz
```

```python
p = pip(xyz)

p
```

```python
%%timeit

p = pip(xyz)
```

```python
p, J_p_xyz = pip_with_grad(xyz)
```

```python
%%timeit

p, J_p_xyz = pip_with_grad(xyz)
```
